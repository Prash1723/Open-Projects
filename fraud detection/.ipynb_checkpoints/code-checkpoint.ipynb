```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'data/house_sales.csv')
```

```python
df.head(5)
```

```python
df.describe()
```

```python
df['digit'] = df['saleprice'].apply(lambda x: int(str(x)[0]))
```

```python
plt.figure(figsize=[12, 8])
count_digit = df['digit'].value_counts().reset_index()
n = df.shape[0]
count_digit['frequency'] = count_digit['count']/n

count_digit['upper_limit'] = count_digit['frequency'].apply(
    lambda x: x + 1.96 * np.sqrt((x*(1+x))/n) + 1/(2*n)
)
count_digit['lower_limit'] = count_digit['frequency'].apply(
    lambda x: x - 1.96 * np.sqrt((x*(1-x))/n) - 1/(2*n)
)

sns.pointplot(x='digit', y='frequency', data=count_digit, color='blue',)
sns.pointplot(x='digit', y='upper_limit', data=count_digit, color = 'green')
sns.pointplot(x='digit', y='lower_limit', data=count_digit, color='red')                                        
plt.show()
```
