```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'data/house_sales.csv')
```

```python
df.head(5)
```

```python
df.describe()
```

```python
df['digit'] = df['saleprice'].apply(lambda x: int(str(x)[0]))
```

```python
plt.figure(figsize=[12, 4])
count_digit = df.groupby('digit').apply(lambda s : pd.Series({
    'frequency': s['digit'].value_counts(),
    'proportion': s['digit'].value_counts(normalize=2),
})).reset_index()
n = df.shape[0]
# count_digit['upper_limit'] = count_digit['proportion'].apply(
#     lambda x: x + 1.96 * np.sqrt(x*(1+x)/n) + 1/2*n
# )
# count_digit['lower_limit'] = count_digit['proportion'].apply(
#     lambda x: x - 1.96 * np.sqrt(x*(1-x)/n) - 1/2*n
# )
sns.pointplot(x='digit', y='proportion', data=count_digit)
sns.pointplot(x='digit', y='upper_limit', data=count_digit)
sns.pointplot(x='digit', y='lower_limit', data=count_digit)                                        
plt.show()
```
